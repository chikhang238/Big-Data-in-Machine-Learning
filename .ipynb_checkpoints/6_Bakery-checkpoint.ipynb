```python
final_data_1 = data_off.withColumn('explode', explode(final_data['items']))
final_data_1.show(5)
```

```python
from pyspark.sql import functions as F
final_data_off = final_data_1.select('id', 'explode').groupBy('id').agg(F.collect_set('explode'))
```

I did this step because FPGrowth just accept the unique items in each record. If I had not done this step, I could not have trained FPGrowth. The special thing is that I used collect_set instead of collect_list

# Use FPGrowth for training

```python
final_data_off.show(5)
```

```python
final_data_off = final_data_off.withColumnRenamed('collect_set(explode)', 'items')
final_data_off.show(5)
```

```python
from pyspark.ml.fpm import FPGrowth


fpGrowth = FPGrowth(itemsCol = 'items', minSupport = 0.005, minConfidence = 0.005)
model = fpGrowth.fit(final_data_off)
```

```python
freq = model.freqItemsets.sort(desc('freq'))
```

```python
freq.show(15, False)
```

```python
df = freq.toPandas()
df.head(10)
```

```python
df['number'] = df['items'].apply(lambda x: len(x) > 1)
```

```python
goods_new = goods.withColumn('flavor_new', regexp_extract(col('Flavor'), '(.)(\w+)(.)', 2))
goods_new = goods_new.withColumn('food_new', regexp_extract(col('Food'), '(.)(\w+)(.)', 2))
```

```python
goods_new.show(5)
```

```python
goods_new.createOrReplaceTempView('goods')

goods_new = spark.sql('SELECT CONCAT(flavor_new, " ", food_new) AS name FROM goods')

goods_new.show(5)
```

```python
good_df = goods_new.toPandas()
good_df.head(5)
```

```python
name = list(good_df['name'])
name
```

```python
apri = pd.read_csv('75000/75000-out2.csv', names = [-1] + name)
apri.head()
```

```python
apri = apri.drop([-1], axis = 1)
```

```python
apri.head()
```

```python
apri = apri.astype(bool)
```

```python
apri.head()
```

```python
from mlxtend.frequent_patterns import apriori
item_frequency = apriori(apri, min_support=0.03, use_colnames = True)
item_frequency.head()
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(item_frequency, metric = 'confidence', min_threshold=0.03)
rules
```

```python
for row in rules.iterrows():
    if 'Chocolate Cake' in row[1][0]:
        print(row)
```

We obtained the result that "Chocolate Cake" is usually bought with "Casino Cake", "Chocolate Coffee"

```python
for row in rules.iterrows():
    if 'Lemon Cake' in row[1][0]:
        print(row)
```
